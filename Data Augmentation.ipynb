```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
```

```python
img=image.load_img(r"C:\Users\om420\OneDrive\Pictures\WhatsApp Image 2026-03-29 at 9.24.18 PM (1).jpeg",target_size=(200,200))
```

```python
type(img)
```

```python
datagen=ImageDataGenerator(
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
    fill_mode='constant'
)
```

```python
img=image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch=img.reshape(1,200,200,3)
```

```python
i=0

for output in datagen.flow(input_batch,batch_size=1,save_to_dir=r'C:\Users\om420\OneDrive\Pictures'):
    i=i+1
    
    if i==2:
        break
```
