```python
import numpy as np           
import pandas as pd                 
import matplotlib.pyplot as plt                       
```

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

# Generating the non-linear dataset shown in the video
X, y = make_moons(n_samples=1000, noise=0.1, random_state=42)

# Visualizing the data
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
import tensorflow as tf            
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

without using activation function

```python
model=Sequential()

model.add(Dense(128,activation='linear',input_dim=2))
model.add(Dense(128,activation='linear'))
model.add(Dense(1,activation='sigmoid'))

adam=Adam(learning_rate=0.1)
model.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])

history=model.fit(X,y,epochs=400,validation_split=0.2,verbose=0)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,y.astype('int'),clf=model,legend=2)
```

Using activation function

```python
model=Sequential()

model.add(Dense(128,activation='relu',input_dim=2))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

adam=Adam(learning_rate=0.1)
model.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])

history=model.fit(X,y,epochs=400,validation_split=0.2,verbose=0)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,y.astype('int'),clf=model,legend=2)
```
